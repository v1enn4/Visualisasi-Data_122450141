Dataset yang digunakan merupakan DNA UTS PBF kelas RC.

Bedasarkan data yang dipilih, tentukan tujuan, user, fungsi dan tone yang sesuai dengan visualisasi yang akan dibuat.

Tujuan : Memvisualisasikan distribusi dari data, melihat frekuensi mahasiswa dari rentang nilai dari data yang tertera
User : Kemungkinan besar user adalah mahasiswa, dosen, atau peneliti yang menganalisis data terkait nilai UTS pada matakuliah PBF.

fungsi : memberikan gambaran yang jelas terkait distribusi data numerik pada kolom DNA, dalam bentuk analisis statistik dasar.

```python
#membaca dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/DNA UTS PBF RC 2024.csv")
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.hist(data['DNA'], bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution dari DNA : ', fontsize=14)
plt.xlabel('Nilai DNA', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

#visualisasi plot
plt.grid(True)
plt.show()
```
